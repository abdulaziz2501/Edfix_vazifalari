### No-odatiy qiymatlarni topish

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/MOCK_USER_DATA.csv", index_col=0)
df.head()
```

Yuqoridagi jadvlada odamlarning yoshi jamlangan. Keling jadval haqida qisqa ma'lumot olamiz:

```python
df.describe()
```

Nimadur xato: `max` yosh `4444` deb ko'rsatayapti, tabiiyku bunday bo'lishi mumkin emas. `min` da ham manfiy qiymat ko'rishimiz mumkin, bu ham xato. Demak jadvalimizda noto'g'ri qiymatlar ham kirib qolgan.

Keling, boshlanishiga katta qiymatlarni filtrlab olamiz, yaqin tarixda odam yoshi maksimum `123` dan oshmagnini bilgan holda, keling shu qiymatdan katta qatorlarni topamiz:

```python
df[df.age>123]
```

Qatorlarning indeksini ham ajratib olishimiz mumkin:

```python
idx = df[df.age>123].index
idx
```

Manfiy qiymatlarni ham yuqoridagi usul bilan filtrlab olishimiz mumkin:

```python
df[df.age<0]
```

Ikki filtrni jamlab yozish ham mumkin:

```python
df[(df.age>123) | (df.age<0)]
```

Qatorlar bilan qanday yo'l tutish sizga bog'liq. Manfiy qiymatlarni musbatga o'tkazishimiz mumkin, buning uchun `numpy` dagi `abs` funksiyasini qo'llaymiz:

```python
df['age'] = df['age'].apply(np.abs)
df.head(15)
```

Katta qiymatlarni tashlab yuboramiz:

```python
idx = df[df.age>123].index
df.drop(index=idx, inplace=True)
df.describe()
```

Ko'rib turganingizdek, jadval qiymatlari normal holatga qaytdi.

Jadvlada yana bir e'tiborimni tortgna jihati, jinslar faqatgina erkak (`Male`) va ayol (`Female`) emas.

```python
df.gender.value_counts()
```

Agar bu sizning keyingi hisoblaringizga ta'sir o'tkazadi desangiz, bu qiymatlarni ham filtrlab olishingiz mumkin:

```python
df[df['gender'].isin(['Male','Female'])]
```

### Jadvlani aralashtirib tashlash

Ba'zida, tadqiqod uchun jadvaldagi qiymatlar o'rnini aralashtirib tashlash talab qilinishi mumkin. Misol uchun, sizda ma'lumotlar ma'lum bir tartibda turibdi, lekin siz ma'lumotlarni tasodifiy tartibda chiqarmoqchiisiz, buning uchun `numpy` kutubxonasining `random` modulidagi `permutation` funksiyasiga murojaat qilamiz.

Bizda quyidagi jadval bor:

```python
df.head()
```

Keling, jadval uzunligini topaylik:

```python
len(df)
```

Jadvalimizda 93 ta qator mavjud, biz 1-93 gacha sonlarni tasodifiy tartibda saqlovchi ro'yxat yaratamiz:

```python
randlist = np.random.permutation(93)
randlist
```

Navbat, `df` ni yuqoridagi tasodifiy tartibda tahlash, buning uchun `.take()` meotidga murojaat qilamiz

```python
df.take(randlist)
```

Marhamat, jadvalimiz tasodifiy tartibda tahlandi.

### Jadvladan  tasodifiy qiymatlarni chiqarib olish.
Buning uchun bizga avvaldan tanish .sample() metodiga murojaat qilamiz va aynan nechta qiymat kerak ekanini ham ko'rsatamiz.

Bunda tasodifiy qatorlar **takrorlanmagan** holda chiqadi.

```python
df.sample(5)
```

Agar qatorlar takrorlanishini istasangiz, `replace=True` parametrini ko'rsatasiz:

```python
df.sample(10, replace=True)
```
